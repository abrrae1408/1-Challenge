# 5.- Envío promedio por tienda

```python
def total_envio(te):
  # Change 'Costo de envio' to the correct column name
  # For example, if the column name is 'Costo De Envío', then use:
  envios_total = te["Costo de envío"].sum()
  return envios_total

def promedio_envios(pe):
  envios_total = total_envio(pe)
  promedio_entrega = envios_total / len(pe)
  return round(promedio_entrega, 2)

print('total de envio de tienda:', total_envio(tienda))
print('promedio de envio de tienda:', promedio_envios(tienda))
print('total de envio de tienda2:', total_envio(tienda2))
print('promedio de envio de tienda:', promedio_envios(tienda2))
print('total de envio de tienda3:', total_envio(tienda3))
print('promedio de envio de tienda:', promedio_envios(tienda3))
print('total de envio de tienda4:', total_envio(tienda4))
print('promedio de envio de tienda:', promedio_envios(tienda4))
```
